### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import numpy as np
import matplotlib.pyplot as plt

# Define x values from 0 to 10
try:
    x = np.linspace(0, 10, 100)
    if x.size == 0:
        raise ValueError("X values array is empty")

    y = x**2  # Square of x

    # Create a line plot
    plt.plot(x, y, label='y = x^2', color='b')

    # Customizations
    plt.title("Line Plot of y = x²")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
