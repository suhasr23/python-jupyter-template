### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

try:
    if any(len(lst) == 0 for lst in [x, y, categories, values, x_scatter, y_scatter, labels, sizes]):
        raise ValueError("One or more data lists are empty")

    # Create 2x2 subplot layout
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # Line chart
    axs[0, 0].plot(x, y, color='blue')
    axs[0, 0].set_title("Line Plot")

    # Bar chart
    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axs[0, 1].set_title("Bar Chart")

    # Scatter plot
    axs[1, 0].scatter(x_scatter, y_scatter, color='orange', marker='o')
    axs[1, 0].set_title("Scatter Plot")

    # Pie chart
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'], startangle=140)
    axs[1, 1].set_title("Pie Chart")

    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
